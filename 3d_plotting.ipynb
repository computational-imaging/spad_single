```python
%matplotlib inline
import torch
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
n =68
X, Y = np.meshgrid(range(n), range(n))
```

```python
Z = np.zeros((n, n))
for i in range(n):
    Z[i, i:] = np.ones(n-i)
```

```python
ax.plot_surface(X, Y, Z, cmap='Blues')
ax.view_init(40,80)
```
